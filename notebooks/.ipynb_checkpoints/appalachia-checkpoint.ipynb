```python
import pandas as pd
import geopandas as gpd
import contextily as ctx
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ARC Counties.csv", dtype={"FIPS": str})
df["FIPS"] = df["FIPS"].str.zfill(5)

counties = gpd.read_file('tl_2022_us_county.zip')
counties["FIPS"] = counties.STATEFP + counties.COUNTYFP
```

```python
app = counties.merge(df, on="FIPS", how="inner")

ax = app.plot(figsize=(10, 8), edgecolor="blue", color="silver", linewidth=0.5)
ax.set_axis_off()

ax = app.to_crs(epsg=3857).plot(figsize=(10, 8), alpha=0.6)
ctx.add_basemap(ax)
ax.set_axis_off()
plt.savefig(
```
